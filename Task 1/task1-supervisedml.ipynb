**Task 1: Prediction using Supervised ML**

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Loading the required Dataset**

```python
df = pd.read_csv('../input/task-1-tsf/student_scores%20-%20student_scores.csv')
```

```python
df.head()
```

```python
df.describe()
```

**Checking for Null Values**

```python
df.isnull().sum()
```

**Plot for the distribution of scores**

```python
plt.scatter(df['Hours'], df['Scores'], color = 'green')
plt.title('Hours vs Percentage(%)')
plt.xlabel('Hours Studied')
plt.ylabel("Percentage Score(%)")
plt.show()
```

**Preparing the data**

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

**Partitioning the data into Training and Testing Sets**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

**Training the model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
print("Model trained!")
```

**Plotting the Regression Line**

```python
line = model.coef_*x + model.intercept_
plt.scatter(x, y, color = 'green')
plt.plot(x, line)
plt.show()
```

**Making Predictions**

```python
print(x_test)
y_pred = model.predict(x_test)
```

**Comparing the Actual and Predicited Scores**

```python
df1 = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred})
df1
```

**Testing with custom input**

```python
hrs=[[5.9]]
predict = model.predict(hrs)
print("No. of Study Hours = {}".format(hrs))
print("Predicted Score={}".format(predict[0]))
```

**Model Evaluation**

```python
from sklearn import metrics
print('Mean Absolute Error(MAE) :', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error(MSE) :', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error(RMSE) :', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
